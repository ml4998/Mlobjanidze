## Clean Census Data

extract census tract from NAME

```python
def extract_tract(tract_string):
    # Split the string by comma
    parts = tract_string.split(',')
    # Get the first part which contains the tract number
    tract_part = parts[0]
    # Split by space and get the last part which is the number
    tract_number = tract_part.split(' ')[-1]
    # convert to string and remove decimal point
    tract_number = str(tract_number).replace('.','')
    return tract_number

census_df['tract_number'] = census_df['NAME'].apply(extract_tract)


```

edit borough name to match evictions data

```python


borough_dict = {
    'Queens County': 'QUEENS',
    'Kings County': 'BROOKLYN',
    'New York County': 'MANHATTAN',
    'Bronx County': 'BRONX',
    'Richmond County': 'STATEN ISLAND'
}

census_df['borough'] = census_df['borough'].map(borough_dict)

census_df
```

```python
# rename the columns to match evictions
census_df = census_df.rename(columns={'tract_number': 'Census Tract', 'borough': 'BOROUGH'})
```

## Merge on Census Tract column

```python
# Assuming 'borough_column' and 'census_tract_column' are the names of the columns
# you want to merge on, replace these with the actual column names from your datasets.
merged_df = pd.merge(eviction_df, census_df, left_on=['BOROUGH', 'Census Tract'], right_on=['BOROUGH', 'Census Tract'], how='inner')
merged_df
```

```python
# output to csv
merged_df.to_csv('merged_data.csv', index=False)
```
